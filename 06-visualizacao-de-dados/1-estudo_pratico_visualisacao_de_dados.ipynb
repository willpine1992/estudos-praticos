# Meu Primeiro Dash App Interativo

Este projeto demonstra a criação de um aplicativo web interativo simples usando a biblioteca Dash para Python, ideal para construir dashboards analíticos. O aplicativo permite visualizar dados de vendas de frutas por cidade de forma dinâmica.

## Funcionalidades

*   **Visualização de Dados:** Exibe um gráfico de barras que compara a quantidade de diferentes frutas (Maçãs, Laranjas, Bananas) em várias cidades.
*   **Interatividade:** Inclui um dropdown que permite ao usuário filtrar o gráfico por cidade, atualizando a visualização em tempo real.
*   **Tecnologias:** Desenvolvido puramente em Python, utilizando:
    *   `Dash`: Framework principal para construção de aplicativos web.
    *   `Plotly Express`: Para a criação de gráficos informativos e esteticamente agradáveis.
    *   `Pandas`: Para manipulação e organização dos dados.

## Como Funciona

O aplicativo carrega um conjunto de dados simulado de vendas de frutas. O layout inicial apresenta um título, uma breve descrição e um gráfico de barras mostrando a distribuição das vendas. Um callback é implementado para conectar um componente de `Dropdown` (seleção de cidade) a um componente `Graph`. Quando uma nova cidade é selecionada no dropdown, o callback filtra os dados e atualiza o gráfico para mostrar apenas as vendas daquela cidade.

````python
import dash
from dash import dcc
from dash import html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd
````

Agora, vamos criar uma amostra de dados que usaremos para o gráfico. Para este exemplo, usaremos um conjunto de dados simples com `x` e `y`.

````python
# df é o dataframe. Neste caso, estamos atribuindo os valores de um df generico dentro do pandas via pd.DataFrame
df = pd.DataFrame({
    "Fruit": ["Apples", "Oranges", "Bananas", "Apples", "Oranges", "Bananas"],
    "Amount": [4, 1, 2, 2, 4, 5],
    "City": ["SF", "SF", "SF", "Montreal", "Montreal", "Montreal"]
})

fig = px.bar(df, x="Fruit", y="Amount", color="City", barmode="group")
````

Em seguida, inicializamos o aplicativo Dash e definimos seu layout. O layout descreve como seu aplicativo será exibido no navegador. Usaremos `html.Div` para estruturar a página e `dcc.Graph` para exibir o gráfico.

````python
app = dash.Dash(__name__)

app.layout = html.Div(children=[
    html.H1(children='Meu Primeiro Dash App'),

    html.Div(children='Um exemplo simples de um aplicativo Dash.'),

    dcc.Graph(
        id='example-graph',
        figure=fig
    )
])
````

Para tornar o aplicativo interativo, precisamos adicionar callbacks. Um callback é uma função Python que é automaticamente chamada por Dash sempre que um `Input` é alterado. Ela pode atualizar as propriedades de um `Output`.

Para este exemplo, criarei um input simples de dropdown para filtrar os dados por cidade. Quando a cidade selecionada no dropdown mudar, o gráfico será atualizado.

**Nota:** Para executar o aplicativo Dash e visualizá-lo, você precisaria executar o `app.run_server(debug=True)` em um ambiente onde o servidor web possa ser acessado, como um script Python ou um ambiente local. Em um notebook Colab, a visualização direta é um pouco diferente, mas você pode usar bibliotecas como `jupyter_dash` para integração mais suave ou usar um túnel para acessar o servidor localmente.

````python
from dash import Dash, dcc, html, Input, Output

# Ao invés de jupyter_dash.JupyterDash, usamos dash.Dash diretamente
app = Dash(__name__)

app.layout = html.Div([
    html.H1("Dash Interativo no Colab"),
    dcc.Dropdown(
        id='city-dropdown',
        options=[
            {'label': i, 'value': i} for i in df['City'].unique()
        ],
        value='SF'
    ),
    dcc.Graph(id='live-update-graph')
])

@app.callback(
    Output('live-update-graph', 'figure'),
    Input('city-dropdown', 'value'))
def update_graph_live(selected_city):
    filtered_df = df[df['City'] == selected_city]
    fig = px.bar(filtered_df, x="Fruit", y="Amount", color="City", barmode="group")
    return fig

# Para rodar o aplicativo no Colab
# Removemos 'mode=inline' para evitar o erro de TypeError com run_simple()
app.run(port=8050, debug=True)
````

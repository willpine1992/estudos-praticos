```python
import os
import pandas as pd
from sqlalchemy import create_engine
```

```python
# ==========================================
# 1. CONFIGURAÇÕES GERAIS
# ==========================================
# Conexão com o PostgreSQL no Docker
db_string = "postgresql://postgres:suasenha123@localhost:5432/banco_cba"
engine = create_engine(db_string)
# Caminhos exatos das duas pastas PROJETOS e NUCLEOS no Windows
caminho_projetos = r"G:\Drives compartilhados\Laboratórios CBA\Gestão de projetos\2026\PROJETO"
caminho_gerenciamento = r"G:\Drives compartilhados\Laboratórios CBA\Gestão de projetos\2026\NUCLEOS"
```

```python
# ==========================================
# 2. FUNÇÃO MOTOR DE ETL (Extração e Carga)
# ==========================================
def processar_e_enviar(caminho_pasta, nome_tabela_sql):
    dados_consolidados = []

    print(f"\n--- Processando pasta: {caminho_pasta} ---")

    # Varre a pasta e subpastas
    for pasta_raiz, diretorios, arquivos in os.walk(caminho_pasta):
        for arquivo in arquivos:
            # Extrai a extensão do arquivo e converte para minúsculo (ex: '.PDF' vira '.pdf')
            extensao = os.path.splitext(arquivo)[1].lower()

            # Trava de Segurança: Só aceita Excel E não pode ser arquivo temporário (~)
            if extensao in ['.xlsx', '.xls'] and not arquivo.startswith('~'):
                caminho_completo = os.path.join(pasta_raiz, arquivo)
                nome_nucleo = os.path.basename(pasta_raiz)

                try:
                    df = pd.read_excel(caminho_completo)
                    df['Nucleo_Origem'] = nome_nucleo
                    df['Arquivo_Origem'] = arquivo
                    dados_consolidados.append(df)
                    print(f"  [OK] Lido: {arquivo}")
                except Exception as e:
                    print(f"  [ERRO] Falha ao ler {arquivo}: {e}")
            else:
                # ADICIONADO: Avisa no console que ignorou um arquivo não autorizado
                print(f"  [IGNORADO] Arquivo não-Excel ou oculto: {arquivo}")

    # Consolida e envia para o SQL
    if dados_consolidados:
        df_final = pd.concat(dados_consolidados, ignore_index=True)
        print(f"Enviando {len(df_final)} linhas para a tabela '{nome_tabela_sql}'...")

        # Substitui a tabela antiga pela nova versão consolidada
        df_final.to_sql(nome_tabela_sql, engine, if_exists='replace', index=False)
        print(f"Sucesso! Tabela '{nome_tabela_sql}' atualizada.")
    else:
        print(f"Aviso: Nenhuma planilha válida encontrada em {caminho_pasta}.")
```

```python
# ==========================================
# 3. EXECUÇÃO DO SCRIPT
# ==========================================
if __name__ == "__main__":
    print("INICIANDO ATUALIZAÇÃO DO DATA WAREHOUSE...")

    # Bloco 1: Atualiza os Projetos
    try:
        processar_e_enviar(caminho_projetos, 'tb_projetos')
    except Exception as e:
        print(f"Erro fatal ao processar Projetos: {e}")

    # Bloco 2: Atualiza o Gerenciamento de Núcleos
    try:
        processar_e_enviar(caminho_gerenciamento, 'tb_gerenciamento')
    except Exception as e:
        print(f"Erro fatal ao processar Gerenciamento: {e}")

    print("\nATUALIZAÇÃO CONCLUÍDA!")
```
